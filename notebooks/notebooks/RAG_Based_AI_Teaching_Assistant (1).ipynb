```python
# converts the videos to mp3
# process_video.py
import os
import subprocess

video_folder = "/content/drive/MyDrive/RAG Based AI Teaching Assistant/Videos"
audio_folder = "/content/drive/MyDrive/RAG Based AI Teaching Assistant/audios"

os.makedirs(audio_folder, exist_ok=True)

files = os.listdir(video_folder)

for file in files:

    tutorial_number = file.split(" ")[-1].split(".")[0]

    file_name = file.split(" ")[0]

    subprocess.run([
        "ffmpeg",
        "-i",
        os.path.join(video_folder, file),
        os.path.join(audio_folder, f"{tutorial_number}_{file_name}.mp3"),
        "-y"
    ])

print("Done!")
```

```python
import os

audio_folder = "/content/drive/MyDrive/RAG Based AI Teaching Assistant/audios"
json_folder = "/content/drive/MyDrive/RAG Based AI Teaching Assistant/jsons"

os.makedirs(json_folder, exist_ok=True)

print("JSON folder ready.")
```

Process all MP3 files and keep in json folder

```python
import os
import json

audios = sorted(os.listdir(audio_folder))

for audio in audios:

    if audio.endswith(".mp3") and "_" in audio:

        print(f"\nProcessing : {audio}")

        number = audio.split("_")[0]
        title = audio.split("_")[1].replace(".mp3", "")

        result = model.transcribe(
            audio=os.path.join(audio_folder, audio),
            task="translate",
            word_timestamps=False
        )

        chunks = []

        for segment in result["segments"]:

            chunks.append({
                "number": number,
                "title": title,
                "start": segment["start"],
                "end": segment["end"],
                "text": segment["text"]
            })

        lecture_data = {
            "number": number,
            "title": title,
            "full_text": result["text"],
            "chunks": chunks
        }

        json_file = os.path.join(
            json_folder,
            audio.replace(".mp3", ".json")
        )

        with open(json_file, "w", encoding="utf-8") as f:
            json.dump(
                lecture_data,
                f,
                indent=4,
                ensure_ascii=False
            )

        print(f"Saved : {json_file}")

print("\nAll lectures completed successfully!")
```

Generate BGE-M3 Embeddings

```python
# Load the BGE-M3 model
from sentence_transformers import SentenceTransformer

model = SentenceTransformer(
    "BAAI/bge-m3",
    device="cuda"     # Uses your T4 GPU
)
```

```python
# Create Embedding Function (create_embedding())
def create_embedding(text_list):
    embeddings = model.encode(
        text_list,
        normalize_embeddings=True,
        convert_to_numpy=True
    )

    return embeddings.tolist()
```

```python
# Test the model
a = create_embedding([
    "Cat sat on the mat",
    "Harry dances on a mat"
])

print(a)
```

Save embeddings.joblib

```python
import os
import json
import pandas as pd
import joblib

# ===============================
# Generate DataFrame from all JSON files
# ===============================

jsons = sorted(os.listdir("jsons"))

my_dicts = []

chunk_id = 0

for json_file in jsons:

    with open(f"jsons/{json_file}", "r", encoding="utf-8") as f:
        content = json.load(f)

    print(f"Creating Embeddings for {json_file}")

    # -----------------------------------
    # Step 1 : Extract all chunk texts
    # -----------------------------------
    texts = [chunk["text"] for chunk in content["chunks"]]

    # -----------------------------------
    # Step 2 : Generate BGE-M3 embeddings
    # create_embedding() internally calls
    # model.encode()
    # -----------------------------------
    embeddings = create_embedding(texts)

    # -----------------------------------
    # Step 3 : Attach embedding to every chunk
    # -----------------------------------
    for i, chunk in enumerate(content["chunks"]):

        chunk["chunk_id"] = chunk_id
        chunk["embedding"] = embeddings[i]

        chunk_id += 1

        my_dicts.append(chunk)

# ===============================
# Create DataFrame
# ===============================

df = pd.DataFrame.from_records(my_dicts)

print(df.head())

# ===============================
# Save DataFrame using Joblib
# ===============================

joblib.dump(df, "embeddings.joblib")

print("\nDataFrame saved successfully!")
```

Generate Final Answer using Gemini

```python
# =======================================================
# Generate Response using Gemini using Inference Function
# =======================================================

def inference(prompt):

    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=prompt
    )

    return response.text
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# =====================================
# Ask Question
# =====================================

incoming_query = input("Ask a Question: ")

# =====================================
# Create Query Embedding
# =====================================

question_embedding = np.array(
    create_embedding([incoming_query])[0]
)

# =====================================
# Compute Cosine Similarity
# =====================================

all_embeddings = np.vstack(df["embedding"].values)

similarities = cosine_similarity(
    all_embeddings,
    question_embedding.reshape(1, -1)
).flatten()

# =====================================
# Retrieve Top 5 Chunks
# =====================================

top_results = 5

max_index = similarities.argsort()[::-1][:top_results]

new_df = df.iloc[max_index]

# =====================================
# Display Retrieved Chunks
# =====================================

print("=" * 60)
print("Retrieved Chunks")
print("=" * 60)

for index, item in new_df.iterrows():

    print(f"\nLecture Number : {item['number']}")
    print(f"Lecture Title  : {item['title']}")
    print(f"Start Time     : {item['start']}")
    print(f"End Time       : {item['end']}")
    print(f"Text           : {item['text']}")
```

```python
# =====================================
# Create Prompt
# =====================================

prompt = f"""
You are an AI Teaching Assistant.

You are given subtitle chunks retrieved from a YouTube course.

Each chunk contains:

- Lecture Number
- Lecture Title
- Start Timestamp
- End Timestamp
- Subtitle Text

Retrieved Chunks:

{new_df[['title','number','start','end','text']].to_json(orient='records', indent=2)}

--------------------------------------------

Student Question:

{incoming_query}

--------------------------------------------

Instructions:

1. Answer ONLY from the retrieved chunks.

2. Explain naturally.

3. Mention the lecture number.

4. Mention the lecture title.

5. Mention the starting timestamp.

6. Mention the ending timestamp.

7. Recommend which lecture to watch.

8. If multiple lectures are relevant, mention all.

9. If the answer is unavailable, clearly say so.

10. Do not hallucinate.
"""

print(prompt)

# =====================================
# Save Prompt
# =====================================

with open("prompt.txt", "w", encoding="utf-8") as f:
    f.write(prompt)

print("\n✅ Prompt saved successfully!")
```

```python
# =====================================
# Generate Response
# =====================================

response = inference(prompt)

print("=" * 60)
print("AI Teaching Assistant")
print("=" * 60)

print(response)

# =====================================
# Save Response
# =====================================

with open("response.txt", "w", encoding="utf-8") as f:
    f.write(response)

print("\n✅ Response saved successfully!")
```
